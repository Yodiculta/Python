```python
# это класс вершины, содержащий её название и названия ближайших доступных для перемещения
class Node(object):  
    def __init__(self, name, array):
        self.name = name
        self.reachable=array
        self.prev=None
        self.num=len(array)
    def set_prev(self, name):
        self.prev=name
    def set_num(self, num):
        self.num=num
```
